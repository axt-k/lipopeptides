```python
import pandas as pd
```

```python
df = pd.read_csv('../cheminformatics/ground_truth_curated.csv')

acc_to_categories = {}

for i in range(len(df)):
    if df.iloc[i]['starter_category'] in ['SCFA', 'MCFA', 'LCFA', 'AR', 'FA']:
        if not df.iloc[i]['mibig_v4_accession'] in  acc_to_categories:
            acc_to_categories[df.iloc[i]['mibig_v4_accession']] = []
        if df.iloc[i]['starter_category'] not in acc_to_categories[df.iloc[i]['mibig_v4_accession']]:
            acc_to_categories[df.iloc[i]['mibig_v4_accession']].append(df.iloc[i]['starter_category'])

for acc, categories in acc_to_categories.items():
    if len(categories) > 1:
        print(acc, categories)
        
```
